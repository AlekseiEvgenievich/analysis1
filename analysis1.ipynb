Определить эффективность маркетинговых каналов по привлечению платящих игроков


```python
retention = df_1.groupby(['subscribing_channel']).agg({"user_id":"count","converted":"sum"}).reset_index()
```

```python
retention['retention_rate'] = retention.converted / retention.user_id
```
